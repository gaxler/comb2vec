```python
from scp_with_or_tools import graph_iterator_from_path, SetCoverSolverFromGraph
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
it = graph_iterator_from_path('/home/gregory/projects/nips18/data/scp/nodes_100_200')
cols = ['items', 'time', 'av_set_size', 'objective']
rows = []
for g in it:
    s = SetCoverSolverFromGraph(graph=g).solve()
    row = [s.items, s.wall_time, s.reduced_set_size, s.objective]
    rows.append(row)
df = pd.DataFrame(data=np.array(rows), columns=cols)
```

```python
sns.set(style='ticks')
sns.pairplot(df, hue="objective")
```
